```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Verificar estilos disponibles
print("Estilos disponibles:", plt.style.available)

# Usar un estilo vlido (por ejemplo, 'ggplot')
plt.style.use('ggplot')

# --- 1. Distribucin Normal ---
def normal_distribution_example():
    mu, sigma = 0, 1  # Media y desviacin estndar
    x = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(x, mu, sigma)

    # Graficar
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label=f'Normal(={mu}, ={sigma**2})', color='blue')
    plt.title('Distribucin Normal')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

# --- 2. Distribucin Exponencial ---
def exponential_distribution_example():
    lambd = 1  # Parmetro lambda
    x = np.linspace(0, 5, 1000)
    pdf = stats.expon.pdf(x, scale=1/lambd)

    # Graficar
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label=f'Exponencial(={lambd})', color='green')
    plt.title('Distribucin Exponencial')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

# --- 3. Distribucin Poisson ---
def poisson_distribution_example():
    lambd = 3  # Parmetro lambda
    x = np.arange(0, 15)
    pmf = stats.poisson.pmf(x, lambd)

    # Graficar
    plt.figure(figsize=(8, 5))
    plt.stem(x, pmf, basefmt=" ", label=f'Poisson(={lambd})')
    plt.title('Distribucin Poisson')
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
    plt.legend()
    plt.show()

# --- 4. Distribucin Bernoulli ---
def bernoulli_distribution_example():
    p = 0.5  # Probabilidad de xito
    x = [0, 1]
    pmf = stats.bernoulli.pmf(x, p)

    # Graficar
    plt.figure(figsize=(8, 5))
    plt.bar(x, pmf, color='purple', alpha=0.7, label=f'Bernoulli(p={p})')
    plt.title('Distribucin Bernoulli')
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
    plt.xticks(x)
    plt.legend()
    plt.show()

# Ejecutar las funciones
normal_distribution_example()
exponential_distribution_example()
poisson_distribution_example()
bernoulli_distribution_example()
```

```python
# Importar libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parmetros de la distribucin Gamma
k = 3       # Parmetro de forma (shape)
theta = 2   # Parmetro de escala (scale)

# Generar valores para el eje x
x = np.linspace(0, 30, 1000)  # Rango de valores en el eje x

# Calcular la funcin de densidad de probabilidad (PDF)
pdf = gamma.pdf(x, a=k, scale=theta)

# Graficar la distribucin
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Gamma(k={k}, ={theta})', color='blue')
plt.fill_between(x, pdf, alpha=0.2, color='blue')  # Sombrear bajo la curva
plt.title('Distribucin Gamma', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Densidad de probabilidad', fontsize=14)
plt.grid(alpha=0.4, linestyle='--')
plt.legend(fontsize=12)
plt.show()

# Generar datos aleatorios simulados de la distribucin Gamma
samples = gamma.rvs(a=k, scale=theta, size=1000)

# Graficar el histograma de los datos simulados
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=30, density=True, alpha=0.7, color='green', edgecolor='black', label='Simulaciones')
plt.plot(x, pdf, label='PDF terica', color='red', linewidth=2)
plt.title('Histograma de datos simulados vs PDF', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Densidad', fontsize=14)
plt.grid(alpha=0.4, linestyle='--')
plt.legend(fontsize=12)
plt.show()
```

```python
# Importar libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parmetros de la distribucin binomial
n = 10  # Nmero de ensayos (lanzamientos de moneda)
p = 0.5  # Probabilidad de xito (obtener cara)

# Generar valores posibles (0 a n)
x = np.arange(0, n + 1)

# Calcular la probabilidad para cada nmero de xitos
probabilidades = binom.pmf(x, n, p)

# Graficar la distribucin
plt.figure(figsize=(8, 6))
plt.bar(x, probabilidades, color='skyblue', edgecolor='black')
plt.title('Distribucin Binomial (n=10, p=0.5)', fontsize=16)
plt.xlabel('Nmero de caras', fontsize=14)
plt.ylabel('Probabilidad', fontsize=14)
plt.xticks(x, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
